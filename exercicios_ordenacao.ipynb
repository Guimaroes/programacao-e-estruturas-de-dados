```python
# ATIVIDADE 1

# Pesquisa binária recursiva
def binary_search(array, value, first, last):
  
  if last >= first:
    middle = (first + last) // 2

    if array[middle] == value:
      return True
    
    elif array[middle] > value:
      return binary_search(array, value, first, middle-1)
        
    else:
      return binary_search(array, value, middle+1, last)
  
  else:
    return False

# Teste
array = [10, 20, 30, 40, 50, 60, 80]
value = 30

result = binary_search(array, value, 0, len(array)-1)

if result :
  print("Encontrou")
else :
  print("Não encontrou")
```

```python
# ATIVIDADE 2

# Le o arquivo e retorna um array com o conteúdo
def read_file_into_array(path):

  file = open(path, "r")

  str_content = file.read().splitlines()

  int_content = [int(string) for string in str_content]

  file.close()
  
  return int_content

# Escreve o array no arquivo
def write_array_into_file(array, path):

  file = open(path, "w")
  
  for item in array:
    file.write("%s\n" % item)

  file.close

# Quicksort
def quicksort(array):

  less = []
  equal = []
  greater = []

  if len(array) > 1:
    pivot = array[0]

    for x in array:
      if x < pivot:
        less.append(x)

      elif x == pivot:
        equal.append(x)

      elif x > pivot:
        greater.append(x)
          
    return quicksort(less) + equal + quicksort(greater)
    
  else:
    return array

# Teste

# Pega os valores do numbers_1.txt, ordena e adiciona no partial_1.txt
array1 = read_file_into_array("/content/drive/MyDrive/exercicios_ordenacao/atividade_2/numbers_1.txt")
sorted1 = quicksort(array1)
write_array_into_file(sorted1,"/content/drive/MyDrive/exercicios_ordenacao/atividade_2/partial_1.txt")

# Pega os valores do numbers_2.txt, ordena e adiciona no partial_2.txt
array2 = read_file_into_array("/content/drive/MyDrive/exercicios_ordenacao/atividade_2/numbers_2.txt")
sorted2 = quicksort(array2)
write_array_into_file(sorted2,"/content/drive/MyDrive/exercicios_ordenacao/atividade_2/partial_2.txt")
```

```python
# ATIVIDADE 3

# Merge Sort
def merge_sort(array):

  if len(array) > 1:

    middle = len(array) // 2
    left_half = array[:middle]
    right_half = array[middle:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = 0
    j = 0
    k = 0

    while i < len(left_half) and j < len(right_half):
      
      if left_half[i] < right_half[j]:
        array[k]=left_half[i]
        
        i = i + 1

      else:
        array[k]=right_half[j]
        
        j = j + 1
      
      k = k + 1

    while i < len(left_half):
      array[k] = left_half[i]
      
      i = i + 1
      k = k + 1

    while j < len(right_half):
      array[k] = right_half[j]
      
      j = j + 1
      k = k + 1

# Teste

# Lê o partial_1.txt e partial_2.txt e adiciona os valores em um único array
sorted1 = read_file_into_array("/content/drive/MyDrive/exercicios_ordenacao/atividade_2/partial_1.txt")
sorted2 = read_file_into_array("/content/drive/MyDrive/exercicios_ordenacao/atividade_2/partial_2.txt")
sorted = sorted1 + sorted2

# Utiliza o Merge Sort para ordenar o array
merge_sort(sorted)

# Adiciona o array ordenado no sorted.txt
write_array_into_file(sorted,"/content/drive/MyDrive/exercicios_ordenacao/atividade_3/sorted.txt")
```
